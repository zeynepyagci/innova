```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from pmdarima.arima import auto_arima
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
```

```python
df = pd.read_csv('train.csv')
```

```python
print(df.head())
print(df.shape)
```

```python
df=df.dropna(axis=1, how='all')
null_values=df.isnull().sum()
print(null_values.sum())
```

```python
df['DATETIME'] = pd.to_datetime(df['DATETIME'])
```

```python
df.sort_values(by='DATETIME', inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(df['DATETIME'], df['CPULOAD'])
plt.xlabel('DATETIME')
plt.ylabel('CPULOAD')
plt.title('Zaman Serisi - CPULOAD')
plt.show()
```

```python
train_data_cihaz_301 = df[df['SERVER'] == 'CIHAZ_301']
train_data_cihaz_305 = df[df['SERVER'] == 'CIHAZ_305']
```

```python
train_data_cihaz_301['DATETIME'] = train_data_cihaz_301['DATETIME'].dt.tz_localize(None)
train_data_cihaz_305['DATETIME'] = train_data_cihaz_305['DATETIME'].dt.tz_localize(None)
```

```python
# Tarih değerlerini unique yap ve grupla
grouped_data_301 = train_data_cihaz_301.groupby(pd.Grouper(key='DATETIME', freq='15min')).mean()

# Ortalama CPULOAD değerini içeren yeni bir sütun ekle
grouped_data_301['Ortalama_CPULOAD'] = grouped_data_301['CPULOAD']

# Sadece ilgili sütunları ve verileri görüntüle
grouped_data_301 = grouped_data_301[['Ortalama_CPULOAD']]

grouped_data_301['DATETIME'] = grouped_data_301.index
grouped_data_301 = grouped_data_301.reset_index(drop=True)
train_data_cihaz_301 = grouped_data_301[['DATETIME', 'Ortalama_CPULOAD']]
print(train_data_cihaz_301)
```

```python
# Tarih değerlerini unique yap ve grupla
grouped_data_305 = train_data_cihaz_305.groupby(pd.Grouper(key='DATETIME', freq='15min')).mean()

# Ortalama CPULOAD değerini içeren yeni bir sütun ekle
grouped_data_305['Ortalama_CPULOAD'] = grouped_data_305['CPULOAD']

# Sadece ilgili sütunları ve verileri görüntüle
grouped_data_305 = grouped_data_305[['Ortalama_CPULOAD']]

grouped_data_305['DATETIME'] = grouped_data_305.index
grouped_data_305 = grouped_data_305.reset_index(drop=True)
train_data_cihaz_305 = grouped_data_305[['DATETIME', 'Ortalama_CPULOAD']]
print(train_data_cihaz_305)


```

```python
#fit ARIMA for cihaz_301
model_301 = ARIMA(train_data_cihaz_301['Ortalama_CPULOAD'], order=(0,1,0)) #0,1,0
model_fit_301 = model_301.fit()
print(model_fit_301.summary())
```

```python
#fit ARIMA for cihaz_305
model_305 = ARIMA(train_data_cihaz_305['Ortalama_CPULOAD'], order=(0,1,0)) #0,1,0
model_fit_305 = model_305.fit()
print(model_fit_305.summary())
```

```python
residuals = model_fit_301.resid[1:]
fig, ax = plt.subplots(1,2)
residuals.plot(title='Residuals', ax=ax[0])
residuals.plot(title='Density', kind='kde', ax=ax[1])
plt.show()
```

```python
residuals = model_fit_305.resid[1:]
fig, ax = plt.subplots(1,2)
residuals.plot(title='Residuals', ax=ax[0])
residuals.plot(title='Density', kind='kde', ax=ax[1])
plt.show()
```

```python
# Tahmin yapma
index_values = pd.date_range(start='2023-05-31 00:00:00', periods=96, freq='15min')
forecast_cihaz_301 = model_fit_301.get_forecast(steps=96).predicted_mean
forecast_cihaz_301 = pd.DataFrame({'DATETIME': index_values, 'Ortalama_CPULOAD': forecast_cihaz_301})
forecast_cihaz_301.reset_index(drop=True, inplace=True)
print(forecast_cihaz_301)
```

```python
df_301 = pd.read_csv('forecast_cihaz_301.csv')

df_305 = pd.read_csv('forecast_cihaz_305.csv')

final_df = pd.concat([df_301, df_305], ignore_index=True)

final_df = final_df.sort_values(by='DATETIME')

final_df.reset_index(drop=True, inplace=True)

final_df.insert(0, 'index', range(len(final_df)))

final_df['index'] = final_df.index

final_df = final_df.drop('DATETIME', axis=1)

final_df = final_df.rename(columns={'Ortalama_CPULOAD': 'CPULOAD'})

final_df.to_csv('final.csv', index=False)
```
